```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
baby_names = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
```

```python
# exploring the data set
baby_names.head(10)
```

```python
#ques1 delete unnamed columns
# deletes Unnamed: 0
del baby_names['Unnamed: 0']
baby_names.head()
```

```python
baby_names['State'].value_counts()
```

```python
# ques 2 distribution of male and female
baby_names['Gender'].value_counts()
```

```python
# 'Year' coloumn is deleted to prevent sum 
del baby_names["Year"]
names = baby_names.groupby("Name").sum()  # Names are grouped together
names.head()
```

```python
# ques 3 top 5 most preferred names
names.sort_values("Count", ascending = 0).head()   # sort from biggest value to lowest
```

```python
# ques 4 median name occurence in the dataset
names[names.Count == names.Count.median()]
```

```python
# ques 5 distribution of male and female born count by states
baby_names.groupby(['State','Gender'])['Gender'].count().unstack()
```
